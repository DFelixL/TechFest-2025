4. Penanganan Null Papua

```python
# 1. Ambil nilai IPM dari Papua & Papua Barat
ipm_papua = df.loc[df['Provinsi'] == 'PAPUA', 'IPM'].values[0]
ipm_papua_barat = df.loc[df['Provinsi'] == 'PAPUA BARAT', 'IPM'].values[0]
avg_ipm_papua = (ipm_papua + ipm_papua_barat) / 2

# 2. Tentukan daftar provinsi baru di Papua
papua_baru = ['PAPUA BARAT DAYA', 'PAPUA PEGUNUNGAN', 'PAPUA SELATAN', 'PAPUA TENGAH']

# 3. Imputasi IPM
df.loc[df['Provinsi'].isin(papua_baru), 'IPM'] = avg_ipm_papua

# 4. Hitung rata-rata jumlah penduduk dari Papua Barat Daya & Papua Tengah
pop_barat_daya = df.loc[df['Provinsi'] == 'PAPUA BARAT DAYA', 'Jumlah Penduduk'].values[0]
pop_tengah = df.loc[df['Provinsi'] == 'PAPUA TENGAH', 'Jumlah Penduduk'].values[0]
avg_pop = (pop_barat_daya + pop_tengah) / 2

# 5. Imputasi jumlah penduduk untuk Papua Pegunungan & Papua Selatan
df.loc[df['Provinsi'] == 'PAPUA PEGUNUNGAN', 'Jumlah Penduduk'] = avg_pop
df.loc[df['Provinsi'] == 'PAPUA SELATAN', 'Jumlah Penduduk'] = avg_pop

# 6. Hitung ulang Anggaran APBN per kapita
df['Anggaran APBN per kapita'] = df['Pagu TKDD'] / df['Jumlah Penduduk']

# 7. Imputasi Persentase Penduduk Miskin dari rata-rata Papua & Papua Barat
poverty_papua = df.loc[df['Provinsi'] == 'PAPUA', 'Persentase Penduduk Miskin'].values[0]
poverty_papua_barat = df.loc[df['Provinsi'] == 'PAPUA BARAT', 'Persentase Penduduk Miskin'].values[0]
avg_poverty = (poverty_papua + poverty_papua_barat) / 2

df.loc[df['Provinsi'].isin(papua_baru), 'Persentase Penduduk Miskin'] = avg_poverty
```

5. Penangan Null Sulawesi

```python
# 1. Daftar provinsi Sulawesi lain sebagai referensi
sulawesi_lain = [
    'SULAWESI SELATAN', 'SULAWESI TENGGARA',
    'SULAWESI TENGAH', 'SULAWESI BARAT', 'GORONTALO'
]

# 2. Hitung rata-rata Pagu TKDD dan Realisasi TKDD dari provinsi tersebut
avg_pagu = df.loc[df['Provinsi'].isin(sulawesi_lain), 'Pagu TKDD'].mean()
avg_realisasi = df.loc[df['Provinsi'].isin(sulawesi_lain), 'Realisasi TKDD'].mean()

# 3. Imputasi ke Sulawesi Utara
df.loc[df['Provinsi'] == 'SULAWESI UTARA', 'Pagu TKDD'] = avg_pagu
df.loc[df['Provinsi'] == 'SULAWESI UTARA', 'Realisasi TKDD'] = avg_realisasi

# Hitung ulang hanya untuk Sulawesi Utara
df.loc[df['Provinsi'] == 'SULAWESI UTARA', 'Anggaran APBN per kapita'] = \
    df.loc[df['Provinsi'] == 'SULAWESI UTARA', 'Pagu TKDD'] / \
    df.loc[df['Provinsi'] == 'SULAWESI UTARA', 'Jumlah Penduduk']
```

## ðŸ“Š **Persentase Realisasi TKDD per Provinsi**

**Deskripsi**
- Visualisasi ini menunjukkan seberapa besar dana TKDD (Transfer ke Daerah dan Dana Desa) yang berhasil direalisasikan dibandingkan dengan pagu yang dialokasikan ke tiap provinsi. Disertai juga klasifikasi kinerja provinsi berdasarkan tingkat realisasinya.

**Tujuan**
- Mengidentifikasi provinsi dengan tingkat penyerapan anggaran tinggi, sedang, dan rendah.
- Menyajikan perbandingan kinerja fiskal antar provinsi secara visual dan informatif.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import Patch

# Hitung persentase realisasi
df_clean["Persentase Realisasi"] = (df_clean["Realisasi TKDD"] / df_clean["Pagu TKDD"]) * 100

# Klasifikasikan kinerja
def klasifikasi(persen):
    if persen >= 95:
        return "Tinggi"
    elif persen >= 80:
        return "Sedang"
    else:
        return "Rendah"

df_clean["Kategori Realisasi"] = df_clean["Persentase Realisasi"].apply(klasifikasi)

# Atur warna berdasarkan kategori
warna_map = {
    "Tinggi": "#2ecc71",   # Hijau
    "Sedang": "#f39c12",   # Oranye
    "Rendah": "#e74c3c"    # Merah
}
df_clean["Warna"] = df_clean["Kategori Realisasi"].map(warna_map)

# Urutkan
df_sorted = df_clean.sort_values("Persentase Realisasi", ascending=False)

# Plot bar chart
plt.figure(figsize=(12, 10))
bars = plt.barh(df_sorted["Provinsi"], df_sorted["Persentase Realisasi"], color=df_sorted["Warna"])

# Tambahkan label nilai
for bar in bars:
    plt.text(bar.get_width() + 0.5, bar.get_y() + bar.get_height()/2,
             f"{bar.get_width():.1f}%", va='center', fontsize=9)

# Tambahkan legenda manual
legend_labels = [
    Patch(color="#2ecc71", label="Tinggi (â‰¥ 95%)"),
    Patch(color="#f39c12", label="Sedang (80â€“94.9%)"),
    Patch(color="#e74c3c", label="Rendah (< 80%)")
]
plt.legend(handles=legend_labels, title="Kategori Realisasi", loc="lower right")

# Label dan judul
plt.xlabel("Persentase Realisasi (%)")
plt.title("Persentase Realisasi TKDD per Provinsi")
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## **Analisis Realisasi TKDD vs IPM**

**Tujuan**

```python
# Perbarui df_final agar mencakup IPM
df_final = df[['Provinsi', 'Jumlah Penduduk', 'Persentase Penduduk Miskin', 'Pagu TKDD', 'Realisasi TKDD', 'IPM']].copy()

# Ganti nama kolom untuk konsistensi
df_final.columns = ['Provinsi', 'jumlah_penduduk', 'persen_miskin', 'pagu', 'realisasi', 'ipm']

# Hitung realisasi persen
df_final['realisasi_persen'] = (df_final['realisasi'] / df_final['pagu']) * 100
```

```python
# Scatter plot setelah perbaikan
plt.figure(figsize=(8,6))
sns.scatterplot(data=df_final, x='ipm', y='realisasi_persen')
sns.regplot(data=df_final, x='ipm', y='realisasi_persen', scatter=False, color='orange')

plt.title('Realisasi TKDD vs IPM')
plt.xlabel('Indeks Pembangunan Manusia (IPM)')
plt.ylabel('Persentase Realisasi TKDD (%)')
plt.grid(True)
plt.tight_layout()
plt.show()
```

ðŸ” **Insight Analisis Realisasi TKDD vs IPM**

1. Scatterplot: Realisasi TKDD (%) vs IPM

- Grafik menunjukkan adanya **korelasi positif** antara IPM dan persentase realisasi TKDD.
- Semakin tinggi **Indeks Pembangunan Manusia (IPM)** suatu provinsi, cenderung semakin tinggi pula **tingkat realisasi anggaran TKDD**.
- Hal ini bisa diartikan bahwa daerah dengan pembangunan manusia yang lebih baik memiliki kapasitas implementasi anggaran yang lebih optimal.
- Namun, terdapat beberapa **outlier**, yaitu daerah dengan IPM sedang atau rendah tetapi realisasi anggarannya tinggi. Ini menunjukkan bahwa IPM bukan satu-satunya faktor yang menentukan efektivitas realisasi anggaran.

- **IPM dan persentase penduduk miskin** memiliki korelasi negatif yang kuat (sekitar -0.81), menunjukkan bahwa semakin tinggi IPM, maka tingkat kemiskinan cenderung menurun.
- **IPM dan realisasi TKDD (%)** menunjukkan korelasi positif sedang (sekitar 0.42), mendukung pola yang terlihat pada scatterplot.
- **Jumlah penduduk dan pagu TKDD** berkorelasi positif (sekitar 0.53), yang logis karena daerah dengan populasi besar cenderung mendapat alokasi anggaran lebih tinggi.
- **Persentase penduduk miskin** memiliki korelasi negatif terhadap **realisasi TKDD (%)** (sekitar -0.49), mengindikasikan bahwa kemiskinan tinggi bisa menjadi hambatan dalam efektivitas penyerapan anggaran.

### 5. Apa saja faktor-faktor yang memengaruhi IPM di tiap provinsi Indonesia pada tahun 2023?
- Korelasi negatif kuat ditemukan antara **persentase penduduk miskin** dan **IPM** (r = **-0.81**), yang berarti semakin tinggi angka kemiskinan, semakin rendah IPM suatu provinsi.
- Faktor lain yang memiliki korelasi positif terhadap IPM adalah **realisasi TKDD (%)** dan **besarnya pagu anggaran**, meskipun tidak sekuat pengaruh kemiskinan.
- Plot multivariat juga menunjukkan bahwa provinsi dengan **IPM rendah** cenderung memiliki **persentase miskin tinggi** dan **realisasi anggaran yang kurang optimal**.
